```python
from apyori import apriori
```

```python
superlist = supermarket.values.tolist() #apriori accepts only list in the algorithm thats why you need to convert to list
```

```python
association_rules = apriori(superlist, min_support = 0.005, min_confidence = 0.2, lift = 2)
rule_list = list(association_rules)
rule_list
```
